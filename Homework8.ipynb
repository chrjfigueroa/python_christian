```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize as fit
```

```python
#1a
df = pd.read_csv('GlobalLandTemperaturesByState.csv')
df = df[["dt","AverageTemperature","State"]]
df = df.sort_values(by='dt')
df = df[605910:]
df = df[df['State'].isin(['Wyoming','Nebraska','South Dakota'])]
display(df)
```

```python
df.reset_index(inplace=True)
df.drop('index', axis=1,inplace=True)
```

```python
#1b
average = df.groupby(['dt'])['AverageTemperature'].mean()
new_df = average.to_frame()
new_df.reset_index(inplace=True)
display(new_df)
```

```python
plt.figure() 
plt.scatter(new_df['dt'],new_df['AverageTemperature'])
plt.xlabel("DATE")
plt.ylabel("AVERAGE TEMPERATURE")
plt.title("Date vs. Average Temperature")
plt.show()
```
