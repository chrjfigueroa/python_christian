```python
import numpy as np
import matplotlib.pyplot as plt

"""
1. How many plots?
"""
#1.
def sinusoidal(x,A=1,w=1):
    y = A*np.sin(w*x)
    return y

#2.
x_data = np.linspace(0,2*np.pi,1000)

#3.
amplitudes = np.array([2,5,7,3,1])
wavelengths = np.array([1,3,5,4,2])

#4.
plt.figure(figsize=(10,6.333))
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("5 Sine functions")
#5.
for i in range(5):
    y_data = sinusoidal(x_data,amplitudes[i],wavelengths[i])
    plt.plot(x_data,y_data,label=f"Function {i+1}")
plt.legend()
plt.show()
```

```python
"""
2. Randomness
"""
x_data = np.random.rand(40) *100
y_data = np.random.rand(40) *100

plt.figure(figsize=(10,10))

plt.plot(x_data,label="x_data", linestyle='--', color="red")
plt.plot(y_data,label="y_data", linewidth = 10,color="orange")

plt.legend()
plt.show()
```
