```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
1: Analyzing Air Quality Data
"""
```

```python
#1.1 Load the Data
df = pd.read_csv('global_air_quality.csv')
print(df)
```

```python
df.rename(columns={'Value':'FactValueNumeric'})
```

```python
#1.2 Make a New Column

#Creating list
PM25_Values =  []
for i in range(9450):
    PM25_Values.append(df['FactValueNumeric'][i])

# Creating new column
df['PM25_Value'] = PM25_Values

#Printing
print(df[['FactValueNumeric','PM25_Value']])
```

```python
#1.3 Calculate Average PM2.5 Concentration

#finding the mean PM25 value of each continent by grouping by parent location
values = df['PM25_Value'].groupby(df['ParentLocation'])
avg_values = values.mean()

#creating list
Average_PM25 = []
for i in range(9450):
    parent_loc = df['ParentLocation'][i]
    Average_PM25.append(avg_values[parent_loc])

#creating column
df['Average_PM25'] = Average_PM25

#Printing
print(df[['Location','ParentLocation','PM25_Value','Average_PM25']])
```

```python
#1.4 Analyze Data
#by printing the mean values for each continent
print(avg_values)
```

```python
"""
We can see that the highest average air pollution occurs in the eastern Mediterranean. This
comes as a shock to me because I always thought the Mediterranean was very clean; I associate large bodies
of water with clean air.
"""
```

```python
#1.5 Save the Processed Data
dataframe = df[['Location','ParentLocation','PM25_Value','Average_PM25']]
dataframe.to_csv('data.csv',index=False)
```
